```python
# Requrie Grad
import torch
x = torch.tensor(10.0, requires_grad=True )
y = torch.tensor(2.0)

y_predict = x*y

print("Predict" , y_predict)
print("Actual",y)
```

```python
# Backward, Grad
x = torch.tensor(10.0, requires_grad=True )
y = torch.tensor(2.0)

y_predict = x*y
loss = (y_predict-y)**2

print("Predict" , y_predict)
print("Actual",y.item())
print("loss",loss.item())

loss.backward()
print("Grad", x.grad.item())
```

# For Gradient


```python
import torch
x = torch.tensor(2.0)
w = torch.tensor(3.0, requires_grad=True )
y = torch.tensor(10.0)
learning_rate = 0.1

# Before Update
y_predict = x*w
print("Prediction:", y_predict.item())
print("Actual", y.item())

loss = (y_predict - y)**2
print('loss',loss.item())
loss.backward() 
print('real "W" before update:',w.item())
print("Grad",w.grad.item())

#  After Update
with torch.no_grad():
    w -= learning_rate * w.grad

w.grad.zero_()
print("After Update 'W' is : ",w.item())



```

Training Loop

```python
import torch
x = torch.tensor([1.,2.,3.,4.])
w = torch.tensor(0.0, requires_grad=True )
y = torch.tensor([2.,4.,6.,8.])
learning_rate = 0.1

epochs = 10
for epoch in range(epochs):
    # step 1 - predicton
    y_predict = w*x

    # step 2 - Loss
    loss = ((y_predict - y)**2).mean()

    #  step 3 - Find
    loss.backward()

    # step 4 - Fix
    with torch.no_grad():
        w -= learning_rate * w.grad
    w.grad.zero_()

    print(f"Epoch {epoch+1}: w = {w.item():.4f}, Loss = {loss.item():.4f}  ")
    
    
```
