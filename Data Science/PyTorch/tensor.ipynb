```python
# project 1
# y = mX + C
import torch
x = torch.tensor([1.,2.,3.])
m = torch.tensor([0.0])
c = torch.tensor([0.0])
y = torch.tensor([3.,1.,2.])

y_pred = m*x + c
print('prediction',y_pred)   # prediction
print('actual',y)   #real
loss = ((y_pred - y) ** 2).mean()

print("Loss:", loss)
```
