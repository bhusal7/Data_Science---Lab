```python
# Create tags column
df['tags'] = (
    df['description'].fillna('') + ' ' +
    df['hashtags'].fillna('') + ' ' +
    df['topic'].fillna('')
)

# Convert to lowercase
df['tags'] = df['tags'].str.lower()


# Remove emojis BEFORE tokenization
df['tags'] = df['tags'].apply(lambda x: emoji.replace_emoji(x, replace=''))


# Tokenize
df['tags'] = df['tags'].apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('english'))

df['tags'] = df['tags'].apply(
    lambda words: [word for word in words if word not in stop_words]
)

# Remove Punctuation
df['tags'] = df['tags'].apply(
    lambda words: [word for word in words if word not in string.punctuation]
)

# Lemmatization
lemmatizer = WordNetLemmatizer() 
df['tags'] = df['tags'].apply(
    lambda words : [lemmatizer.lemmatize(word) for word in words]
)
# Join tokens back into text
df['tags'] = df['tags'].apply(lambda words: " ".join(words)) 

df[['tags']].head()
```

Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['tags'])
tfidf_matrix
```

Cosine Similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

# -------------------------------
# Reset Index
# -------------------------------
df = df.reset_index(drop=True)

# -------------------------------
# Cosine Similarity Matrix
# -------------------------------
cosine_sim = cosine_similarity(tfidf_matrix)

# -------------------------------
# Recommendation Function
# -------------------------------

def recommend(topic, n=10):

    # Topic exists?
    if topic not in df["topic"].unique():
        return "Reel not found"

    # Get ALL rows belonging to that topic
    topic_indices = df[df["topic"] == topic].index.tolist()

    # Average similarity of all reels in that topic
    avg_similarity = cosine_sim[topic_indices].mean(axis=0)

    # Remove same-topic reels
    avg_similarity[topic_indices] = -1

    # Top recommendations
    recommended_indices = np.argsort(avg_similarity)[::-1][:n]

    return df.loc[
        recommended_indices,
        [
            "description",
            "account_id",
            "topic",
            "likes",
            "comments"
        ]
    ].reset_index(drop=True)


# -------------------------------
# Test
# -------------------------------

recommend("Trends")
```
